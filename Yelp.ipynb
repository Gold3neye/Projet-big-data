```
from pyspark.ml.feature import StringIndexer 
from pyspark.sql.types import *
from pyspark.sql import functions as F

#ajout colonne user_Id
table_reco = StringIndexer(inputCol="user_name", outputCol="user_Id").fit(reviews_enriched).transform(reviews_enriched)

#ajout colonne place_Id
table_reco = StringIndexer(inputCol="name", outputCol="place_Id").fit(table_reco).transform(table_reco)

#sélection des colonnes à utiliser pour reco
table_reco=table_reco.select(F.col("user_Id"), F.col("place_Id"), F.col("user_rating").cast(IntegerType()))

display(table_reco)
```

```
from pyspark.ml.recommendation import ALS
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

train, test = table_reco.randomSplit([0.7, 0.3])

als = ALS(userCol="user_Id", itemCol="place_Id", ratingCol="user_rating", coldStartStrategy="drop", nonnegative=True)

evaluator = RegressionEvaluator(metricName="rmse", labelCol="user_rating", predictionCol="prediction")

paramGrid = ParamGridBuilder()\
                    .addGrid(als.rank, [1, 5])\
                    .addGrid(als.maxIter, [2, 10])\
                    .build()
#cross validation
cv = CrossValidator(
  estimator=als, 
  evaluator=evaluator, 
  estimatorParamMaps=paramGrid, 
  numFolds=5
)

model = cv.fit(train)

```

```
predictions = model.transform(test)
evaluator.evaluate(predictions)
#predictions = predictions.withColumn("predictionR", F.abs(F.round(predictions["prediction"],0)))
display(predictions)
```
